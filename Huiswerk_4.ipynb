## Hoeveel clusters heb je nodig?

Voor veel clusteringalgoritmes moet je het aantal clusters wat je algoritme gaat identificeren opgeven. Bij weer andere moet je waarden opgeven die bijvoorbeeld de grenswaarde voor dichtheid van punten aangeeft waarboven samples in hetzelfde cluster zitten. Kortom, er zit altijd, naast het gebruikte algoritme, enige arbitraire keuze in hyperparameters wanneer je met clustering bezig bent.

Er is echter wel een soort vuistregel die je houvast biedt bij het kiezen hiervan. De vuistregel die we hier onderzoeken heet de "elleboogmethode". Als je KMeans hebt gefit, dan kent het model-object een attribuut "inertia_" en dat is de som van de afstanden van alle punten in het sample tot hun dichtsbijzijnde clustercentrum. Door toevoeging van meer clusters wordt dit natuurlijk altijd kleiner, maar als je de waarde van de inertia plot tegen het aantal clusters zie je altijd hetzelfde gedrag: in het begin neemt de inertia sterk af als je een cluster toevoegt, en nadat je er "genoeg hebt" wordt die afname veel langzamer. In een plotje van de inertia tegen het aantal clusters zie je dus vaak een "elleboog". De waarde van het aantal clusters waarbij dit optreedt is een goede keuze voor het aantal clusters dat je wil gebruiken. 

Hieronder worden weer twee groepjes blobs gedefinieerd. Gebruik KMeans en het inertia_ attribuut om te onderzoeken hoeveel clusters aangeraden zouden zijn. Vergelijk dit met de creatie van de blobs. Conclusies?

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=200, centers=4, random_state=42)
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1], c=y, s=20)
X2, y2 = make_blobs(random_state=101, centers=[[-8, -8],[0, 0], [2,2]], n_samples=1000, )
plt.subplot(122)
plt.scatter(X2[:, 0], X2[:, 1], c=y2, s=20);
```
